```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
df = pd.DataFrame({
    'Names': ['Omar', 'Fahad', 'Khalid', 'Majed'],
    'Age': [20, 27, 30, 45],
    'Salary':[8000,6490,7390,4567],
    'Gender': ['m','m','m','m']
})
lst1 = ['Nourh', 'Reem', 'Alia', 'Lama']
lst2 = [9000,7800,6700,2384]

#The Female Salary
plt.plot(lst1, lst2, c = 'pink')
#The male Salary
plt.plot(df['Names'],df['Salary'] , c = 'Blue')

plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Employee Salaries');
```

## Q2: Create pie chart with equal explode values.

```python

flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint Chip', 'Cookies and Cream']
scoops = [35, 28, 20, 15, 12]
colors = ['#FDFD96', '#8B4513', '#fe2c54', '#00FA9A', '#eee0b1']

plt.pie(scoops,labels=flavors, autopct='%1.1f%%', startangle=0, colors=colors, shadow=True, radius=1, explode = [0.3,0.2,0,0,0]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python


countries = ['France', 'Italy', 'Japan', 'Australia', 'Spain']
votes = [75, 68, 82, 60, 72]

plt.figure(figsize=[8, 6])
plt.scatter(countries, votes, marker='o', alpha=0.4, s=50)
plt.title('Best Vacation Destinations')
plt.xlabel('Countries')
plt.ylabel('Votes');



```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(figsize=[8, 6])
plt.scatter(countries, votes, marker='o', alpha=0.4, s=50)
plt.title('Best Vacation Destinations', color='red', loc='left', fontsize=18)
plt.xlabel('Countries', fontsize=16)
plt.ylabel('Votes', fontsize=16);
```

## Q5: Create three line charts with Legends at left and colors.

```python

# Using the same data in Q1
df = pd.DataFrame({
    'Names': ['Omar', 'Fahad', 'Khalid', 'Majed'],
    'Age': [20, 27, 30, 45],
    'Salary': [8000, 6490, 7390, 4567],
    'Gender': ['m', 'm', 'm', 'm'],
    'Commission':[1500,2000,1800,3000],
    'Salary deduction':[1000,500,2500,800]
}) 

plt.figure(figsize=[8, 6])

# The line for worked hours per month
plt.plot(df['Names'], df['Salary deduction'], marker='o', color='red', label='Salary deduction')

# The line for Salaries
plt.plot(df['Names'], df['Salary'], marker='o', color='green', label='Salary')

# The line for commissions
plt.plot(df['Names'], df['Commission'], marker='o', color='blue', label='Commission')

# Legends
plt.legend(loc='upper left')

# Set title
plt.xlabel('Countries')
plt.ylabel('Name')
plt.title('Employees Earning');


```

## Q6: Create subplots with three charts and figsize=20,10.

```python

countries = ['France', 'Italy', 'Japan', 'Australia', 'Spain']
votes = [75, 68, 82, 60, 72]
income = [500000, 400000, 600000, 350000, 450000]

plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

# First subplot - Pie chart
plt.subplot(1, 3, 1)
plt.pie(votes, labels=countries, autopct='%1.1f%%', startangle=0, shadow=True, radius=1)
plt.title('Most Voted Countries')

# Second subplot - Scatter plot
plt.subplot(1, 3, 2)
plt.scatter(votes, income)
plt.title('Income vs. Votes')
plt.xlabel('Votes')
plt.ylabel('Income')

# Third subplot - Bar chart
plt.subplot(1, 3, 3)
plt.bar(countries, income)
plt.title('Income by Country')
plt.xlabel('Countries')
plt.ylabel('Income');

```
